```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate data: 9 time snapshots of a 1D quantity
x = np.linspace(0, 10, 100)
data = [np.sin(x + i) for i in range(9)]  # e.g., quantity changing over time

fig, axes = plt.subplots(3, 3, figsize=(9, 9),
                         sharex=True, sharey=True,
                         gridspec_kw={'wspace': 0.1, 'hspace': 0.1})

# Flatten axes for easy iteration
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.plot(x, data[i], color='black')
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(-1.1, 1.1)

# Optional: Add titles or time labels inside plots
for i, ax in enumerate(axes):
    ax.text(0.05, 0.9, f"t={i}", transform=ax.transAxes, fontsize=10, ha='left', va='top')



plt.tight_layout()
plt.show()
```
