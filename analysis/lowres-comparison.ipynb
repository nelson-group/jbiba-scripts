```python
from measurments import Sim, map_unstructured_to_structured_3d_optimized
import numpy as np 
import matplotlib.pyplot as plt
```

```python
lowres = Sim("/u/jbiba/projects/turbulent-driving/driving64-mach0.5")
highres = Sim("/u/jbiba/projects/turbulent-driving/driving128-mach0.5")
```

```python
fig, ax = plt.subplots()
line_sol, gauss_sol = lowres.property_pdf("Density", 350, fig, ax, avg=100)
line_mix, gauss_mix = highres.property_pdf("Density", 350, fig, ax, avg=100)
ax.set_xlabel(r"$\log(\rho/\langle\rho\rangle)$")
ax.set_ylabel("Volume weighted PDF")
```

```python
gauss_sol[0].set_alpha(0.5)
gauss_mix[0].set_alpha(0.5)
gauss_sol[0].set_color("orange")
gauss_mix[0].set_color("green")
line_sol[0].set_color("orange")
line_mix[0].set_color("green")
line_sol[0].set_label(r"$64^3$")
line_mix[0].set_label(r"128^3")
ax.legend()
fig.savefig("density_pdf_64-128_comparison_snap350-average.pdf", bbox_inches="tight")
```
