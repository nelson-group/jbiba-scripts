```python
galaxy = nicegalaxy.Galaxy(nicegalaxy.galaxies[30])
```

```python
mmw = 4/(1 + 3*0.76 + 4*galaxy.gas["ElectronAbundance"]*0.76) * 1.67262192e-24
temperature = (5/3 - 1) * galaxy.gas["InternalEnergy"]/(1.380649e-16) * 1e10 * mmw
```

```python
vols = galaxy.gas["Masses"]/galaxy.gas["Density"]
np.log10(np.sum(temperature*vols)/np.sum(vols))
```

```python
from scipy.constants import Boltzmann as k_B_cgs
from scipy.constants import proton_mass as m_p_si

# Constants
gamma = 5 / 3
X_H = 0.76
k_B = k_B_cgs * 1e7  # Boltzmann constant in CGS (erg/K)
m_p = m_p_si * 1e3   # Proton mass in grams

# Code units (example values - adjust as needed)
UnitLength_in_kpc = 1.0
UnitTime_in_Gyr = 1.0
UnitMass = 1.0       # in 10^10 Msun
UnitEnergy = (UnitLength_in_kpc * 3.086e21)**2 / (UnitTime_in_Gyr * 3.154e16)**2

# Internal energy per unit mass in code units (array)
u_code = galaxy.gas["InternalEnergy"]  # replace with actual snapshot values
x_e = galaxy.gas["ElectronAbundance"]    # ElectronAbundance field

# Mean molecular weight (in grams)
mu = 4.0 / (1.0 + 3 * X_H + 4 * X_H * x_e) * m_p

# Convert temperature (in Kelvin)
T = (gamma - 1) * u_code * (UnitEnergy / UnitMass) * mu / k_B
```

```python
np.log10(np.sum(T*vols)/np.sum(vols))
```
